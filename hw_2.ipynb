#นำเข้า lib

```python
import pandas as pd
```

#อ่านไฟล์

```python
df = pd.read_excel("week02_workshop.xlsx")
```

#สร้าง dic ไว้เก็บชื่อ

```python
name_arry = {}
```

#สร้าง dic ไว้เก็บนามสกุล

```python
last = {}
```

#เอาข้อมูลชื่อทั้งหมด

```python
df_name = df.ชื่อ
```

#เอาข้อมูลนามสกุลทั้งหมด

```python
df_last_name = df.นามสกุล
```

#fuc ไว้ใช้กับชื่อ

```python
def bio(j):
    for i in range(len(j)-1):
        arr_name = j[i:i+2]
        if arr_name in name_arry.keys():
            name_arry[arr_name] = name_arry[arr_name]+1
        else:
            name_arry[arr_name] = 1
    return(name_arry)
```

#ตัวแปร = ข้อมูลที่เราเลือก

```python
name = df_name
```

```python
la = df_last_name
```

#ลูปข้อมูลใช้ fuc และใช้ lower ใช้ตัวเล็กหมด

```python
for i in name:
    bio_name = bio(i.lower())
```

#fuc ไว้ใช้กับนามสกุล

```python
def last_name(j):
    for i in range(len(j)-1):
        last_name = j[i:i+2]
        if last_name in last.keys():
            last[last_name] = last[last_name]+1
        else:
            last[last_name] = 1
    return(last)
```

#ลูปข้อมูลใช้ fuc และใช้ lower ใช้ตัวเล็กหมด

```python
for m in la:
    bio_last = last_name(m.lower())
```
